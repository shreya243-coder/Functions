6. Explain the concept of generators in Python and how they are defined.
-> A generator is a special type of function that returns values one at a time using `yield`, instead of returning all at once.
  How to Define a Generator?  
âœ… Use `def` like a normal function.  
âœ… Use `yield` instead of `return` to produce values.  
  Example:  
```python
def count_up(n):
    for i in range(1, n+1):
        yield i  # Returns one value at a time

gen = count_up(3)  
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
```
  Why Use Generators?  
âœ” Memory efficient (doesn't store all values at once).  
âœ” Used for large data processing (like reading big files).  
âœ” Can be paused and resumed using `yield` and `next()`.

9. Explain the purpose and usage of the `map()` function in Python.
->  Purpose and Usage of `map()` in Python:

The `map()` function applies a given function to all items in an iterable (like a list) and returns a map object (an iterator)
  Syntax:
```python
map(function, iterable)
```

function: The function to apply.
iterable: The iterable (like a list, tuple, etc.) whose elements will be passed to the function.

Example
```python

    return x * x

numbers = [1, 2, 3, 4]

```
  Why Use `map()`?
âœ… Efficient: Applies a function to each item without using loops.
âœ… Readable: Makes code cleaner and more concise.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
-> Difference Between `map()`, `reduce()`, and `filter()`

i. `map()`-
   - Purpose: Applies a function to **every item** in an iterable (e.g., list) and returns a new iterable with the results.
   -Example: Squaring each number in a list.  
   ```python
   numbers = [1, 2, 3]
   squared = map(lambda x: x * x, numbers)
   print(list(squared))  # Output: [1, 4, 9]
   ```

ii. `reduce()`
   - Purpose: Applies a function cumulatively to the items in an iterable (from left to right), reducing them to a single value.  
   - Note: Needs to be imported from `functools`.  
   - Example: Finding the product of all numbers.  
   ```python
   from functools import reduce
   numbers = [1, 2, 3, 4]
   product = reduce(lambda x, y: x * y, numbers)
   print(product)  # Output: 24
   ```

iii. `filter()`  
   -Purpose: Filters the items in an iterable based on a condition (function) and returns only the items that satisfy the condition.  
   - Example: Filtering out even numbers.  
   ```python
   numbers = [1, 2, 3, 4]
   even_numbers = filter(lambda x: x % 2 == 0, numbers)
   print(list(even_numbers))  # Output: [2, 4]
   ```

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
'''
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # Total terms to generate
        self.a, self.b = 0, 1   # First two Fibonacci numbers
        self.count = 0          # Counter to track iterations

    def __iter__(self):
        return self  # The class itself is an iterator

    def __next__(self):
        if self.count >= self.n_terms:  # Stop iteration after n_terms
            raise StopIteration

        fib_number = self.a  # Get the current Fibonacci number
        self.a, self.b = self.b, self.a + self.b  # Update values for next iteration
        self.count += 1  # Increase the counter
        return fib_number

# Example usage:
n = 10  # Number of Fibonacci terms
fib_iterator = FibonacciIterator(n)

# Iterate through the Fibonacci sequence
for num in fib_iterator:
    print(num, end=" ")

'''
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
'''
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent  # Yield 2 raised to the current exponent

# Example usage:
max_exp = 5  # Generate powers of 2 up to 2^5
for power in powers_of_two(max_exp):
    print(power, end=" ")

'''
````

````python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
'''
def read_file_line_by_line(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line.strip()  # Yield each line after stripping extra spaces

# Example usage:
file_path = "example.txt"  # Replace with your actual file path
for line in read_file_line_by_line(file_path):
    print(line)

'''
````
